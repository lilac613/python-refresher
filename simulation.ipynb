```python
import numpy as np
import physics
(times,x,y,theta,v,omega,a) = physics.simulate_auv2_motion(np.array([10,0,10,0]),np.pi/4,8,6,100,100,3)
physics.plot_auv2_motion(times,x,y,theta,v,omega,a)
```

## Parameters

```python
density_water = 1000  # kg/m^3
density_object = 500  # kg/m^3
object_volume = 0.1  # m^3
g = 9.81  # m/s^2
time_step = 0.01  # s
time_max = 10  # s
```

```python
time = 0  # s
depth = 0  # m
velocity = 0  # m/s
acceleration = 0  # m/s^2
buoyancy =   density_water*object_volume*g
weight =   density_object*object_volume*g
```

```python
# lists for plotting later
time = np.arange(0, time_max, time_step)
position = np.zeros_like(time)
velocity = np.zeros_like(time)
acceleration = np.zeros_like(time)
force = np.zeros_like(time)
```

```python
# simulation loop
for i in range(1, len(time)):
    # calculate the force on the object
    net_force = buoyancy - weight

    # calculate acceleration
    acceleration[i] = net_force/(density_object*object_volume)

    # calculate velocity
    velocity[i] = velocity[i-1] + acceleration[i-1]*time_step

    # calculate position
    position[i] = position[i-1] + velocity[i-1]*time_step
```

# Plot the results

```python

import matplotlib.pyplot as plt
```

```python

plt.plot(time, position, label="Position")
plt.plot(time, velocity, label="Velocity")
plt.plot(time, acceleration, label="Acceleration")
plt.xlabel("Time (s)")
plt.ylabel("Position (m), Velocity (m/s), Acceleration (m/s^2)")
plt.legend()
plt.show()
```
